# • Broadcasting in Numpy:-

```python
## The term broadcasting describes how Numpy treats arrays with different shapes during arithmetic operations.

## The Smaller array is 'broadcast' across the larger array so that they have compatible shapes.
```

```python
# Same shape: 

a = np.arange(6).reshape(2, 3)
b = np.arange(6, 12).reshape(2, 3)

print(a)
print(b)

print(a + b)
```

```python
# different shape:

a = np.arange(6).reshape(2, 3)
b = np.arange(3).reshape(1, 3)

print(a)
print(b)

print(a + b)
```

# • Broadcasting Rules :-


```python
"""
1. Make the two arrays have the same number of dimensions.
    • if the number of dimensions of the two arrays are different, add new dimensions with size 1 to the head
    of the array with the smaller dimension.

2. Make each dimension of the two arrays the same size.
    • if the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the 
    size of the other array.

    • if there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an
    error is raised.
    
"""
```

```python
# More Example :-

a = np.arange(12).reshape(4, 3)
b = np.arange(3)

print(a)
print(b)
print()
print(a + b) # broadcasting
```

```python
a = np.arange(12).reshape(3, 4)
b = np.arange(3)

print(a)
print(b)


# print(a + b) # error bcoz: row and col is not going to be same as other 2-D array in broadcasting...
```

# Long Vs Wide Data Format :-

• Wide Format :is where we have a single row for every data point with multiple columns to hold the values of  
    various attributes. 

• Long Format :is where, for each data point we have as many rows as the number of attributes and each row contains
    the value of a particular attribute for a given data point.

```python
# melt -> simple example branch, 
# Convert Wide data format into Long data format.

# Example :-

pd.DataFrame({'cse' : [120]}).melt()
```

```python
# melt -> branch with year 
# Example :-

pd.DataFrame({'cse':[120], 'ece':[100], 'mech':[50]}).melt(var_name='branch', value_name='students')
```

```python
pd.DataFrame(
    {
        'branch': ['cse', 'ece', 'mech'],
        '2020': [100, 150, 60],
        '2021': [120, 130 , 80],
        '2022': [150, 140, 90]
    }
).melt(id_vars=['branch'], var_name='year', value_name='student')
```

# Pivot Tables :-

• The pivot table takes simple columns-wise data as input, and groups the entries into a two-dimensional that provides a multidimensional summarization of the data.

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.groupby('sex')[['total_bill']].mean()
```

```python
df.groupby(['sex', 'smoker'])[['total_bill']].mean().unstack()
```

```python
# pivot table Using :-

df.pivot_table(index='sex', columns='smoker',values='total_bill')
```

```python
# aggfunc :-

df.pivot_table(index='sex', columns='smoker', values='total_bill', aggfunc='sum')
```
