```python
# Class : group  all variables/attributes and functions/methods into single logical unit
import random
import string

class Short_URL:
    def __init__(self): # Constructor: not must , but, good practice to initialize all attributes
        self.d = dict()

    # URLPrefix = "https://www.shortURL.com/"  # Class variable shared by all objects

    # given a long url, get a short url
    def get_ShortURL(self, long_url): # first argument to all methods is 'self' => this object 
        # length = random value in 6-10
        l = random.randint(6,10)

        # generate random characters into a string of length 1
        chars = string.ascii_lowercase
        Short_URL = ''.join(random.choices(chars) for i in range(l))

        # Check if this string is already present in dictionary 'd'
        if Short_URL in self.d:
            return self.get_ShortURL(long_url)
        else:
            self.d[Short_URL] = long_url

        res = "https://www.short.url/" + Short_URL
        return res

    # given a short url, get the long url
    def get_Long_url(self, short_url):
        # Extract the key from the short URL
        key = short_url[25:]
        print(key) 

        if key in self.d:
            return self.d[key]
        else:
            return None

print(Short_URL().get_Long_url("https://www.shortURL.com/mxzmuis"))
```

```python
# Class: datatype/DS & Object : variables of a class
s = Short_URL() # Constructor being called , memory allocated.

print(type(s)) # <class '__main__.Short_URL'>

print(s.get_Long_url("https://www.shortURL.com/mxzmuis"))
```

```python
# Examples: -
# Design a MyURLShortener  built on top of ShortURLFinal (given in module) with some Changes
#   - Changes getShortUrl to have both alphabets and numbers
#   - Change the URLPrefix to my website (myurlshortener.com)
```

```python
class ShortURLFinal:
    URLPrefix = "https://www.shortURL.com/"  # Class variable

    def __init__(self):
        self.d = dict()

    def getShortURL(self, longURL):
        length = random.randint(6, 10)
        chars = string.ascii_lowercase
        Short_URL = ''.join(random.choices(chars) for _ in range(length))

        if Short_URL in self.d:
            return self.getShortURL(longURL)
        else:
            self.d[Short_URL] = longURL

        return self.URLPrefix + Short_URL

    def getLongURL(self, shortURL):
        key = shortURL[len(self.URLPrefix):]
        if key in self.d:
            return self.d[key]
        else:
            return None
```

#

```python
# Lets define MyURLShortener based on shortURLFinal
# Inheritance : base class - parent class/super class , derived class - child class
import random
import string 

class MyURLShortener(ShortURLFinal): # Inheriting from ShortURLFinal

    URLPrefix = "https://www.myurlshortener.com/"  # Overriding as same name as base class

    # Given a long URL, get a Short URL
    def getShortURL(self, longURL):   # Overriding as same name as the base class method, use both digit and lowercase
        # length = random value in [6-10]
        length = random.randint(6,10)

        # Generate a random characters into a string of length 1
        Chars = string.ascii_lowercase + string.digits  # both number and lowercase
        Short_URL = ''.join(random.choice(Chars) for i in range(length))

        # checking this string is already present in dictionary 'd'
        if Short_URL in self.d:
            return self.getShortURL(longURL)
        else:
            self.d[Short_URL] = longURL

        return self.URLPrefix + Short_URL
```

```python
m1 = MyURLShortener() # base class Constructor being called (executed)

print(m1.d)

print(m1.getShortURL("https://www.google.com/"))
print(m1.getShortURL("google.com"))
print(m1.getShortURL("amazon.com"))
```

```python
# Example form geometry;
import math

class Shape:
    def __init__(self, color= 'black', filled= False):
        self.__color = color
        self.__filled = filled

    def get_Color(self):
        return self.__color

    def isFilled(self):
        return self.__filled

    def set_Color(self, color):
        self.__color = color

    def get_Filled(self):
        return self.__filled

    def set_Filled(self, filled):
        self.__filled = filled



class Rectangle(Shape):
    def __init__(self, length, breadth):
        super().__init__()
        self.__length = length
        self.__breadth = breadth

    def get_length(self):
        return self.__length
    
    def set_length(self, length):
        self.__length = length

    def get_breadth(self):
        return self.__breadth
    
    def set_breadth(self, breadth):
        self.__breadth = breadth

    def get_Area(self):
        return self.__length * self.__breadth

    def get_Perimeter(self):
        return 2 * (self.__length + self.__breadth)
    


class Circle(Shape):
    def __init__(self, radius):
        super().__init__()
        self.__radius = radius

    def get_radius(self):
        return self.__radius
    
    def set_radius(self, radius):
        self.__radius = radius

    def get_Area(self):
        return math.pi * self.__radius * self.__radius

    def get_Circumference(self):
        return 2 * math.pi * self.__radius
    

r1 = Rectangle(10.5, 2.5)

print(r1.get_Area())
print(r1.get_Perimeter())
print(r1.get_Color())
print(r1.get_Filled())
r1.set_Filled(True)

print(r1.get_Filled())
r1.set_Color("Red")
print(r1.get_Color())


c1 = Circle(12)

print(f"Area: {c1.get_Area():.2f}")
print(f"Circumference: {c1.get_Circumference():.2f}")
print(f"Color: {c1.get_Color()}")
print(f"Filled: {c1.get_Filled()}")
c1.set_Filled(True)

print(f"Filled: {c1.get_Filled()}")
c1.set_Color("Blue")
print(f"Color: {c1.get_Color()}")
```
